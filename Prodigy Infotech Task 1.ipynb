```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
diamonds_data = sns.load_dataset("diamonds")
```

```python
# Create a bar chart to visualize the distribution of diamond cut
plt.figure(figsize=(8, 6))
sns.countplot(x='cut', data=diamonds_data)
plt.xlabel('Diamond Cut')
plt.ylabel('Count')
plt.title('Distribution of Diamond Cut')
plt.xticks(rotation=45)  # Rotate x-axis labels for better visibility
plt.show()
```

```python
# Create a histogram to visualize the distribution of diamond carat weight
plt.figure(figsize=(8, 6))
sns.histplot(data=diamonds_data, x='carat', bins=30, kde=True)
plt.xlabel('Diamond Carat Weight')
plt.ylabel('Frequency')
plt.title('Distribution of Diamond Carat Weight')
plt.show()
```
